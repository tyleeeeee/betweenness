```python
import json
import numpy as np 
import csv
```

```python
with open('starwars-full-interactions-allCharacters.json', 'r') as j:
    file = json.load(j)
node = file['nodes']
link = file['links']
N = len(node)
L = len(link)
```

adjcent matrix

```python
#initial adj matrix = 0
adj = np.zeros(shape=(N,N),dtype=int) 

#if there is a link between node m and node n, a_mn=1,a_nm=1
for i in range(L): 
    if link[i]['value'] >= 1:
        m = link[i]['source']
        n = link[i]['target']
        adj[m,n] = 1
        adj[n,m] = 1

#æŠŠadjå¯«å…¥matrix.csv
with open('matrix.csv', 'w', newline='') as file: 
    mywriter = csv.writer(file, delimiter=',')
    mywriter.writerows(adj)
```

shortest path

```python
class node:
    def __init__(self,val,path,dist):
        self.val = val
        self.path = path
        self.dist = dist
        
def bfs(start, end):
    #store all nodes to nodes shortest path routes
    paths = []
    #initial all distance = infinite, source distance = 0
    distance = [float('inf')]* N # n is number of nodes
    mindist = float('inf')
    distance[start] = 0
    
    queue = []
    queue.append(node(start,[start],0))
    a = []

    while(queue):
        s = queue.pop(0)
        if(s.val == end):
            mindist = s.dist
            a.append(s)
        if(s.dist > mindist):
            pass
        elif(s.dist <= mindist):
            for i in range(N):
                if(adj[s.val][i] == 1 and distance[i] >= s.dist + 1):
                    queue.append(node(i,s.path + [i],s.dist + 1))
                    distance[i] = s.dist + 1
    for p in a:
        paths.append(p.path)
        
    return(paths)
```

```python
all_path = []  #store all the paths from s to t
for s in range(N):
    sol = []
    for t in range(N):
        if s != t:
            k = bfs(s, t)
            sol.append(k)
        else:
            sol.append([])
            
    all_path.append(sol)
```

betweenness(v)=âˆ‘(ğ‘ â‰ ğ‘¡â‰ ğ‘£)(ğœ_ğ‘ ğ‘¡(ğ‘£))/ğœ_ğ‘ ğ‘¡ 

```python
 #all shortest path from s to t
def st(s,t):
    _st = len(all_path[s][t])
    return _st

 #all shortest path from s to t which pass through v
def st_v(s,t,v):
    _st_v = 0
    for i in range(len(all_path[s][t])):
        c = all_path[s][t][i].count(v)
        _st_v += c
    return(_st_v)
```

```python
bet_v = np.zeros(N) #store all nodes' betweenness
for v in range(N):
    for s in range(N):
        if (v == s):
            continue
        for t in range(N):
            c = 0
            if (t == v or t == s) :
                continue
            if st(s,t) == 0:
                continue
            else:
                c = st_v(s,t,v) / st(s,t)
            bet_v[v] += c
        
bet_v = np.round(bet_v, 2)
```

```python
bet_rank = np.argsort(bet_v)[::-1][:10]
rank = np.zeros(shape=(10,2)) 

for i in range(10): 
    rank[i][0] = bet_rank[i]
    rank[i][1] = bet_v[bet_rank[i]]

#æŠŠrankå¯«å…¥rank.csv
with open('rank.csv', 'w', newline='') as file: 
    mywriter = csv.writer(file, delimiter=',')
    mywriter.writerows(rank)
```
